## What product sold the most & why?

```python
data.columns
```

```python
count_df = data.groupby(['Product']).agg({'Quantity Ordered' : 'sum', 'Price Each' : 'mean'})
```

```python
count_df
```

```python
count_df = count_df.reset_index()
```

```python
count_df
```

```python
product = count_df['Product'].values
```

```python


fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1.bar(count_df['Product'], count_df['Quantity Ordered'])
ax2.plot(count_df['Product'], count_df['Price Each'])
ax1.set_xticklabels(product, rotation='vertical', fontsize=10)
ax1.set_ylabel('Order Count')
ax2.set_ylabel('Avg price of the product')
```

## Analysing what products are most often sold together?

```python
data.columns
```

```python
data['Order ID']
```

```python
data['Order ID'].duplicated(keep=False)
```

```python
df_duplicated = data[data['Order ID'].duplicated(keep=False)]
```

```python
dup_products = df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x : ','.join(x)).reset_index().rename(columns={'Product' : 'grouped_products'})
```

```python
dup_products
```

```python
dup_products_df = df_duplicated.merge(dup_products, how='left', on='Order ID')
```

```python
dup_products_df
```

```python
dup_df_nil = dup_products_df.drop_duplicates(subset=['Order ID'])
```

```python
dup_df_nil
```

```python
dup_df_nil.shape
```

```python
dup_df_nil['grouped_products'].value_counts()
```

```python
dup_df_nil['grouped_products'].value_counts()[0:5].plot.pie()
```
